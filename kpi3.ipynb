```python
# Excluir las rutas 5 y 6
rutas_excluir = [5, 6]

data_filtrada = df[~df['FK_RUTA'].isin(rutas_excluir)]
# Configurar el tamaño de la figura
plt.figure(figsize=(8, 10))

# Crear el boxplot con las rutas filtradas
sns.boxplot(x='FK_RUTA', y='DISTANCIA', data=data_filtrada)

# Añadir título y etiquetas
plt.title("Distribución de Distancias por Ruta")
plt.xlabel("Ruta")
plt.ylabel("Distancia Recorrida")
```

Distancias por debajo de 15000 indican que el viaje se canceló o que el conductor tomó atajos para completar el viaje, por lo cual se eliminan estos registros.

```python
data_filtrada = df[df['DISTANCIA'] >= 15000].copy()

# Mostrar el número de registros antes y después
print(f"Datos originales: {df.shape[0]} registros")
print(f"Datos filtrados: {data_filtrada.shape[0]} registros")
```

```python
#Calcular el MAD (Desviación Absoluta de la Mediana)
def filtrar_outliers_por_ruta(grupo):
    mediana = grupo.median()
    mad = np.median(np.abs(grupo - mediana))
    lim_sup = mediana + 3 * mad
    lim_inf = max(mediana - 3 * mad, 0)  # Evitar negativos
    return grupo[(grupo >= lim_inf) & (grupo <= lim_sup)]

df_clean = df.groupby('FK_RUTA')['DISTANCIA'].apply(filtrar_outliers_por_ruta).reset_index()
#Quedan los registros que tienen una distancia lógica
df_clean
```

```python
# Excluir las rutas 5 y 6
rutas_excluir = [5, 6]

df_clean = df_clean[~df_clean['FK_RUTA'].isin(rutas_excluir)]
# Configurar el tamaño de la figura
plt.figure(figsize=(8, 10))

# Crear el boxplot con las rutas filtradas
sns.boxplot(x='FK_RUTA', y='DISTANCIA', data=df_clean)

# Añadir título y etiquetas
plt.title("Distribución de Distancias por Ruta")
plt.xlabel("Ruta")
plt.ylabel("Distancia Recorrida")
```

Por regla de negocio, que los viajes que cumplieron el recorrido son aquellos que su distancia está 5% por encima o por debajo de la mediana

```python
# Calcular la mediana de la distancia por ruta
mediana_por_ruta = df_clean.groupby('FK_RUTA')['DISTANCIA'].median()
mediana_por_ruta
```

```python
# Calcular la media de la distancia por ruta
media_por_ruta = df_clean.groupby('FK_RUTA')['DISTANCIA'].mean()
media_por_ruta
```

```python
# Crear una nueva columna con la mediana correspondiente a cada registro
data_filtrada = df_clean.copy()
data_filtrada['mediana'] = data_filtrada['FK_RUTA'].map(mediana_por_ruta)

# Definir los límites de aceptación (5% por encima y por debajo de la mediana)
data_filtrada['limite_inferior'] = data_filtrada['mediana'] * 0.95
data_filtrada['limite_superior'] = data_filtrada['mediana'] * 1.05

# Filtrar los datos dentro del rango aceptable
data_resultante = data_filtrada[(data_filtrada['DISTANCIA'] >= data_filtrada['limite_inferior']) & (data_filtrada['DISTANCIA'] <= data_filtrada['limite_superior'])]
data_resultante
```

```python
# Excluir las rutas 5 y 6
rutas_excluir = [5, 6]

data_resultante = data_resultante[~data_resultante['FK_RUTA'].isin(rutas_excluir)]
# Configurar el tamaño de la figura
plt.figure(figsize=(8, 10))

# Crear el boxplot con las rutas filtradas
sns.boxplot(x='FK_RUTA', y='DISTANCIA', data=data_resultante)

# Añadir título y etiquetas
plt.title("Distribución de Distancias por Ruta")
plt.xlabel("Ruta")
plt.ylabel("Distancia Recorrida")
```

```python
# Contar registros por ruta de aquellos que cumplen con la distancia
conteo_por_ruta = data_resultante.groupby('FK_RUTA').size().reset_index(name='total_registros_por_ruta') #crea un dataframe
conteo_por_ruta
```

```python
# Contar registros por ruta de todos los registros resultantes (cumplen y no cumplen)
conteo_por_ruta_general = df_clean.groupby('FK_RUTA').size().reset_index(name='total_registros_por_ruta_general') #crea un dataframe
conteo_por_ruta_general
```

% cumplimiento en las distancias de duración de los viajes

```python
# Unir ambos DataFrames en base a 'FK_RUTA'
porcentaje_cumplimiento = conteo_por_ruta.merge(conteo_por_ruta_general, on='FK_RUTA')
porcentaje_cumplimiento
```

```python
porcentaje_cumplimiento['%_cumplimiento'] = (
    porcentaje_cumplimiento['total_registros_por_ruta'] /
    porcentaje_cumplimiento['total_registros_por_ruta_general']
) * 100
porcentaje_cumplimiento
```
